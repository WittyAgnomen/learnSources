```python
# cleaning (needs to be more robust)
text = df['article'][0]
text = text.replace('\n', '').replace('....','.').replace('Õ','').replace('Ó','').replace('Ò','').replace('.', ' ').replace('!', ' ').replace('?', ' ').replace(',','')
text = text.split(' ')
```

```python
text
```

```python
re.findall("(<[a-z]{1,2}>)$","Trump")==None
```

```python
tokens = []
flag = False
for word in text:
    x = re.findall("<[a-z]{1,2}>", word.lower())
    if x != []:
        tokens.append((word, x[0]))
    else:
        tokens.append((word, "None"))
```
